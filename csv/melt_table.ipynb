```python
import pandas as pd
df_copy = pd.read_csv('data.csv')
df = df_copy.copy()
```

```python
df
```

```python
df_retantion = df[['customerkey', 'cohort_year', 'm_1', 'm_3', 'm_6', 'm_12']].copy()
```

```python
df_retantion
```

```python
df_melted = df_retantion.melt(id_vars=['customerkey', 'cohort_year'], value_vars=df_retantion.columns[1:], value_name='retirned_customer', var_name='month')
```

```python
df_melted['month'] = df_melted['month'].str.replace('m_', 'Month ')
```

```python
df_melted
```

```python
df_melted.to_csv('cohort_retention.csv', sep=';', index=True, encoding='utf-8-sig')
```
